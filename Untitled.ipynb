```python
import pandas as pd, pickle
from pathlib import Path
import sys
sys.path.insert(0, './MotifBreakdown')
sys.path.insert(0, './r-BRICS')
import generate_vocab_rules as gvr
from rdkit import Chem, RDLogger
RDLogger.DisableLog('rdApp.*')

smi = 'Nc1ccc(N=Nc2ccc(N=Nc3c(S(=O)(=O)O)cc4cc(S(=O)(=O)O)c(N=Nc5ccc(N=Nc6ccc(N)cc6N)cc5)c(O)c4c3N)cc2)c(N)c1'
mol = Chem.MolFromSmiles(smi)

pieces = gvr.fragment_molecule_tracked(mol, smi, use_fallback=False, method='rbrics_old')

vdir = Path('vocab_output/Mutagenicity/rbrics_old')
cols = pd.read_csv(vdir / 'matrix_columns.csv')
frag_to_id = dict(zip(cols['motif_identity'], cols['motif_id']))

# Simulate build_lookup for this one molecule
node_map = {}
for smarts, atom_set in pieces:
    print(f"smarts in frag_to_id: {smarts in frag_to_id}")
    print(f"smarts: {smarts[:80]}")
    mid = frag_to_id.get(smarts, 'MISSING')
    print(f"mid: {mid}")
    for atom_idx in atom_set:
        node_map[atom_idx] = (smarts, mid)

print(f"node_map entries: {len(node_map)}")
```

```python
import sys, copy
sys.path.insert(0, './MotifBreakdown')
sys.path.insert(0, './r-BRICS')
from rdkit import Chem, RDLogger
RDLogger.DisableLog('rdApp.*')
import molfragbpe5 as frag, pandas as pd

df = pd.read_csv('/nfs/hpc/share/kokatea/ChemIntuit/MotifBreakdown/datasets/FOLDS/Mutagenicity_0.csv')
nitro = '[*][N+](=O)[O-]'

rb_has, af_has = [], []
for smi in df['smiles']:
    mol = Chem.MolFromSmiles(smi)
    if mol is None: continue
    
    frag._CACHE.clear()
    h1 = frag.Hierarchy()
    rb = frag.fragment_molecule(mol, h1, use_fallback=False, method='rbrics')
    if nitro in rb: rb_has.append(smi)
    
    frag._CACHE.clear()
    h2 = frag.Hierarchy()
    af = frag.fragment_molecule(mol, h2, use_fallback=True, method='all')
    if nitro in af: af_has.append(smi)

rb_only = set(rb_has) - set(af_has)
print(f"rBRICS has nitro: {len(rb_has)}")
print(f"all+fallback has nitro: {len(af_has)}")
print(f"Lost by all+fallback: {len(rb_only)}")
for s in list(rb_only)[:3]:
    mol = Chem.MolFromSmiles(s)
    frag._CACHE.clear()
    h = frag.Hierarchy()
    print(f"\n  {s}")
    print(f"  all+fb: {frag.fragment_molecule(mol,h,use_fallback=True,method='all')}")
```

```python
import sys, copy
sys.path.insert(0, './MotifBreakdown')
sys.path.insert(0, './r-BRICS')
from rdkit import Chem, RDLogger
RDLogger.DisableLog('rdApp.*')
import molfragbpe5 as frag, pandas as pd

df = pd.read_csv('/nfs/hpc/share/kokatea/ChemIntuit/MotifBreakdown/datasets/FOLDS/Mutagenicity_0.csv')
tv = df[df['group'].isin(['training','valid'])]
nitro = '[*][N+](=O)[O-]'

print(f"Total train+val molecules: {len(tv)}")

rb_count, af_count = 0, 0
for smi in tv['smiles']:
    mol = Chem.MolFromSmiles(smi)
    if mol is None: continue
    
    frag._CACHE.clear()
    h1 = frag.Hierarchy()
    rb = frag.fragment_molecule(mol, h1, use_fallback=False, method='rbrics')
    if nitro in rb: rb_count += 1
    
    frag._CACHE.clear()
    h2 = frag.Hierarchy()
    af = frag.fragment_molecule(mol, h2, use_fallback=True, method='all')
    if nitro in af: af_count += 1

print(f"rBRICS: {rb_count} molecules with nitro → {rb_count/len(tv)*100:.1f}%")
print(f"all+fb: {af_count} molecules with nitro → {af_count/len(tv)*100:.1f}%")
```
